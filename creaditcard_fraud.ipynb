```python
from sklearn.model_selection import train_test_split
```

```python
# get col names for features and target
features= df.iloc[:, 1:30].columns
target= df.iloc[:1, 30:]
print("targets:\n", target, 'features:\n', features)
df
```

```python
# feature df
feature_df= df[features]
# tareget df
target_df= df.Class

```

```python
X_train, X_test, y_train, y_test= train_test_split(feature_df, target_df, train_size=.7, test_size=.3, random_state=1)

print("len of x-train is: {x}".format(x=len(X_train)))
print("len of x-test: {x}".format(x=len(X_test)))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
y_train
```

```python
model= LogisticRegression()
model.fit(X_train, y_train.values.ravel())

pred= model.predict(X_test)
pred
```

```python

matrix = confusion_matrix(y_test, pred)
matrix

df_matrix= pd.DataFrame(matrix, columns= ["no_fraud", 'fraud'], index=  ["no_fraud", 'fraud'])
sns.heatmap(df_matrix, annot=True, cbar=None, cmap="Blues")
plt.title("Confusion Martix")
```

```python

# F1- Score reprsents a more balanced result as it is the harmonic mean between Precision and Recall.
from sklearn.metrics import f1_score, recall_score
f1_score = round(f1_score(y_test, pred), 2)
f1_score
recall_score = round(recall_score(y_test, pred), 2)
recall_score
print("Sensitivity/Recall for Logistic Regression Model 1 : {recall_score}".format(recall_score = recall_score))
print("F1 Score for Logistic Regression Model 1 : {f1_score}".format(f1_score = f1_score))
```
